# Classification

```python
import pycaret.classification as cl
```

```python
s = cl.setup(df, target = 'Churn')
```

```python
best = cl.compare_models()
```

- Accuracy: a quel points ce que j'ai dit bons sont bien bons et ceux que j'ai dis mauvais sont bien mauvais
- Precision: a quel point ce que j'ai dit bons sont bien bons
- Recall: quel pourcentage de ceux qui sont vraiment bons j'ai réussi à trouver
- F1: Mixe entre Recall et Accuracy
- AUC: aire sous la courbe ROC

```python
print(best)
```

```python
cl.evaluate_model(best)
```

# Clustering

```python
import pycaret.clustering as clus
```

```python
s = clus.setup(df, normalize = True)
```

```python
kmeans = clus.create_model('kmeans')
```

#### Silhouete:

a: The mean distance between a sample and all other points in the same class.
b: The mean distance between a sample and all other points in the next nearest cluster.

The Silhouette Coefficient s for a single sample is then given as: (b-a)/(max(a,b))

The score is higher when clusters are dense and well separated, which relates to a standard concept of a cluster.
The Silhouette Coefficient is generally higher for convex clusters than other concepts of clusters, such as density based clusters like those obtained through DBSCAN.


#### Calinski-Harabasz Index

The Calinski-Harabasz index also known as the Variance Ratio Criterion - can be used to evaluate the model, where a higher Calinski-Harabasz score relates to a model with better defined clusters.

The index is the ratio of the sum of between-clusters dispersion and of within-cluster dispersion for all clusters (where dispersion is defined as the sum of distances squared):

```python
clus.evaluate_model(kmeans)
```
